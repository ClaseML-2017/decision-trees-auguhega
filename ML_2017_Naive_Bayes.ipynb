```python
import numpy as np
import pandas as pd 
import csv
import random as rnd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
df=pd.read_csv('spambase.data', header=None)
```

```python
 X_train, X_test, Y_train, Y_test = train_test_split(df.ix[:,0:56],df[[57]], train_size=0.75)
```

```python
X_train=np.array(X_train)
X_test=np.array(X_test)
Y_train=np.array(Y_train)
Y_test=np.array(Y_test)
```

Probabilidad de SPAM y NO SPAM

```python
X_train[np.array(Y_train==1)]
```

```python
Prob_spam=float(sum(Y_train)/len(Y_train))
```

```python
Prob_nospam=1-Prob_spam
```

Calculamos medias y desviaciones estandar de las variables cuando son spam y cuando no lo son

```python
mean_s=np.mean(X_train[np.array(Y_train==1)], axis=0)
desv_s=np.std(X_train[np.array(Y_train==1)], axis=0)
mean_nos=np.mean(X_train[np.array(Y_train==0)], axis=0)
desv_nos=np.std(X_train[np.array(Y_train==0)], axis=0)
```

```python
isspam=X_train[np.array(Y_train==1)]
notspam=X_train[np.array(Y_train==0)]
```

```python
def SPAMPred (x):
    spam= np.log(Prob_spam)
    nospam=np.log(Prob_nospam)
    calc_spam=0
    calc_nspam=0
    
    for i in range(len(isspam)):
        if mean_s[i]!=0 and desv_s[i]>0:
            calc_spam=calc_spam+np.log(np.random.normal(mean_s[i], desv_s[i]))
            
    for i in range(len(notspam)):
        if mean_nos[i]!=0 and desv_nos[i]>0:
            calc_nspam=calc_nspam+np.log(np.random.normal(mean_nos[i], desv_nos[i]))
            
    if (spam+calc_spam) > (nospam+calc_nospam):
        return 1
    else:
        return 0
    
```

```python
prediction= [SPAMPred(x) for x in np.array(X_test)]
```
