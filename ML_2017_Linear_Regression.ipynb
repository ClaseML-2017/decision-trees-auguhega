# Regresion Lineal 

```python
import pandas as pd
import numpy as np
import csv
import random as rnd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
import matplotlib.pyplot as plt
%matplotlib inline
```

Leemos los datos de la regresion lineal

```python
df=pd.read_csv('regLin.csv')
```

Dividimos los datos en datps de entrenamiento y validacion en proporcion 75/25

```python
 X_train, X_test, Y_train, Y_test = train_test_split(df[['X']],df[['y']], train_size=0.75)
```

Graficamos los datos de entrenamiento

```python
plt.scatter(X_train, Y_train)
```

```python
plt.scatter(X_test, Y_test)
```

Corremos una regresión con los datos de entrenamiento

```python
linReg=LinearRegression()
linReg.fit(X_train, Y_train)
```

La regresion nos arroja el intercepto y el coeficiente

```python
print linReg.intercept_, linReg.coef_[0]
```

Obtenemos el error con la diferencia al cuadrado del y-predecida y y-original usando los datos de validacion

```python
np.mean((linReg.predict(X_test)-Y_test)**2)
```

```python
x=np.linspace(float(X_test.min()),float(X_test.max()),100)
```

Graficamos los datos de validacion con la linea de prediccion obtenida con la ecuacion de la regresion

```python
plt.scatter(X_test,Y_test)
plt.plot(x*linReg.coef_[0]+linReg.intercept_,color='red')
```

```python
pred=(X_test*linReg.coef_[0]+linReg.intercept_)
pred2=linReg.predict(X_test)
```

```python
plt.scatter(X_test, pred2)
```

```python
error=pred2-Y_test
plt.plot(error['y'].tolist())
```

```python
#calculamos una serie de valores para wo y w1
wo = np.linspace(linReg.intercept_-5, linReg.intercept_+5, 50)
w1 = np.linspace(linReg.coef_[0]-3, linReg.coef_[0]+3,50)
```

```python
y=np.asarray(Y_test)
x=np.asarray(X_test)
y.shape
```

```python
F1=np.zeros(50)
F2=np.zeros(50)
```

```python
np.mean((y-linReg.intercept_-linReg.coef_[0]*x)**2)
```

```python
#Ejemplificamos como cambia el error cuando cambia wo
for i in range(50):
    F1[i]=np.mean((y-wo[i]-linReg.coef_[0]*x)**2)
```

```python
plt.scatter(wo,F1)
```

```python
#Ejemplificamos como cambia el error cuando cambia w1
for i in range(50):
    F2[i]=np.mean((y-linReg.intercept_-w1[i]*x)**2)
```

```python
plt.scatter(w1,F2)
```
